```python
import pandas as pd
import numpy as np

st_loc = pd.read_csv('station_coordinate.csv', encoding='utf8')
df = pd.read_csv('Total_Bicycle.csv', encoding='utf8')
```

```python
df['대여일시']
df['대여일'] = df['대여일시'].str.split(" ", expand=True)[0]
df['대여시간'] = df['대여일시'].str.split(" ", expand=True)[1]
df['대여시간'] = df['대여시간'].str.split(":", expand=True)[0]

from datetime import datetime, date

def what_day(date):
    days = ['월', '화', '수', '목', '금', '토', '일']
    day = date.weekday()
    return days[day]
    
dday = []
for i in range(len(df)):
    qday = df['대여일'][i]
    datetime_date = datetime.strptime(qday, '%Y-%m-%d')
    qday = what_day(datetime_date)
    dday.append(qday)
df['대여요일'] = dday

df['반납일시']
df['반납일'] = df['반납일시'].str.split(" ", expand=True)[0]
df['반납시간'] = df['반납일시'].str.split(" ", expand=True)[1]
df['반납시간'] = df['반납시간'].str.split(":", expand=True)[0]
    
dday = []
for i in range(len(df)):
    qday = df['반납일'][i]
    datetime_date = datetime.strptime(qday, '%Y-%m-%d')
    qday = what_day(datetime_date)
    dday.append(qday)
df['반납요일'] = dday

df2 = df
```

```python
df3 = df2[10000:10300] #예시용으로 300개만 읽게 함.
#df3 = df2
df3.reset_index(drop=True,inplace=True)

df_diff_rental = pd.DataFrame()
df_diff_return = pd.DataFrame()
station = []
diff_rental = []
diff_return = []
df3_rank1_station_rental = []
df3_rank1_diff_rental = []
df3_rank2_station_rental = []
df3_rank2_diff_rental = []
df3_rank1_station_return = []
df3_rank1_diff_return = []
df3_rank2_station_return = []
df3_rank2_diff_return = []
ex = pd.DataFrame()

for i in range(len(df3)):
    for j in range(len(st_loc)):
        station.append(st_loc['name'][j])
        diff_rental.append(np.sqrt((df3['대여소 위도'][i]-st_loc['lat'][j])**2 +
                            (df3['대여소 경도'][i]-st_loc['lng'][j])**2))
        diff_return.append(np.sqrt((df3['반납소 위도'][i]-st_loc['lat'][j])**2 +
                            (df3['반납소 경도'][i]-st_loc['lng'][j])**2))
    ex['역'] = station
    ex['대여'] = diff_rental
    ex['반납'] = diff_return
    df_diff_rental['역(대여)'] = station
    df_diff_return['역(반납)'] = station
    df_diff_rental['차이(대여)'] = diff_rental
    df_diff_return['차이(반납)'] = diff_return
    df_diff_rental.sort_values('차이(대여)', inplace=True)
    df3_rank1_station_rental.append(df_diff_rental.iloc[0]['역(대여)'])
    df3_rank1_diff_rental.append(df_diff_rental.iloc[0]['차이(대여)'])
    df3_rank2_station_rental.append(df_diff_rental.iloc[1]['역(대여)'])
    df3_rank2_diff_rental.append(df_diff_rental.iloc[1]['차이(대여)'])
    df_diff_return.sort_values('차이(반납)', inplace=True)
    df3_rank1_station_return.append(df_diff_return.iloc[0]['역(반납)'])
    df3_rank1_diff_return.append(df_diff_return.iloc[0]['차이(반납)'])
    df3_rank2_station_return.append(df_diff_return.iloc[1]['역(반납)'])
    df3_rank2_diff_return.append(df_diff_return.iloc[1]['차이(반납)'])
    station = []
    diff_rental = []
    diff_return = []
    
df3['1순위(대여)'] = df3_rank1_station_rental
df3['1순위 차이(대여)'] = df3_rank1_diff_rental
df3['2순위(대여)'] = df3_rank2_station_rental
df3['2순위 차이(대여)'] = df3_rank2_diff_rental
df3['1순위(반납)'] = df3_rank1_station_return
df3['1순위 차이(반납)'] = df3_rank1_diff_return
df3['2순위(반납)'] = df3_rank2_station_return
df3['2순위 차이(반납)'] = df3_rank2_diff_return
df3
```
