```python
import pandas as pd

bicycle_data = pd.read_csv("C://Users//gwcat//Downloads//서울특별시 공공자전거 대여이력 정보_22.06.csv", encoding='CP949')
bicycle_data.head(5)

```

```python
# 전처리 해야할 일 

# (대여 대여소 = 반납대여소) & (이용시간 = 0 or 이용거리 =0) 인 행 삭제 -> 해결
# [서울 전체] 대여소번호, 대여소명 → ㅇㅇ동으로 replace -> 해결했는데 동매칭 안 되는 20개 있어 일일히 매핑 필요 
# 자전거번호 대신 고유한 index 추가 -> 해결
# 대여 이용자 수/지역 히트맵  & 반납 이용자 수/ 지역 히트맵
```

```python
# 필요없는 컬럼 삭제

del bicycle_data['반납거치대']
```

```python
# 데이터의 결측치 확인하기 
bicycle_data.isnull().sum()

# 결측치는 없는 것을 확인 !
```

```python
# 1. 대여소와 반납소가 같은 사용자 중에 이용시간 또는 이용거리가 0인 행 삭제하기
# 1) 삭제할 데이터 확인하기
bicycle_data[(bicycle_data['대여 대여소번호']==bicycle_data['반납대여소번호']) & ( (bicycle_data['이용시간']==0) | (bicycle_data['이용거리']==0.0)) ]
```

```python
# 2)삭제할 데이터의 갯수 세보기 
bicycle_data[(bicycle_data['대여 대여소번호']==bicycle_data['반납대여소번호']) & ( (bicycle_data['이용시간']==0) | (bicycle_data['이용거리']==0.0)) ].count()

# 갯수를 세보니 15003개 데이터가 필요없는 데이터로 확인되었음.
```

```python
# 3) 삭제하기 
index1 = bicycle_data[(bicycle_data['대여 대여소번호']==bicycle_data['반납대여소번호']) & ( (bicycle_data['이용시간']==0) | (bicycle_data['이용거리']==0.0)) ].index
bicycle_data=bicycle_data.drop(index1)
```

```python
# 데이터 크기(갯수) 확인
bicycle_data.shape

# 총 4181982 데이터로 15003개가 삭제되었음을 알 수 있다. -> 삭제완료 🧡
```

```python
# 2. 자전거번호 대신 고유한 index 추가
bicycle_data.reset_index(inplace=True)
bicycle_data.head(5)
```

```python
# 좌표->주소 변환
from geopy.geocoders import Nominatim

def geocoding_reverse(lat_lng_str): 
    geolocoder = Nominatim(user_agent = 'South Korea', timeout=None)
    address = geolocoder.reverse(lat_lng_str)

    return address

address = geocoding_reverse('37.55495071,126.9108353')
address2 = geocoding_reverse('37.52464294,126.9234238')

print(address)
```

```python
rantalshop_data['dong'] = None
```

```python
# 위도 경도로 주소(제대로 된)와 동 들고오기
for i, j in rantalshop_data.iterrows():

    print(i)
    g = geocoding_reverse((j['Unnamed: 4'], j['Unnamed: 5']))
    

    rantalshop_data['address'][i] = g[0]
    print(g[0])
    
    
    a_list = g[0].split(',')
    for c in range(1,len(a_list)):
        if a_list[c][-1] == '구' :
            if a_list[c-1][-1] == '동':
                dong = a_list[c-1].strip()
            else:
                dong ='없음'

    print('동:',dong)
    rantalshop_data['dong'][i] = dong

```
