```python
# 2)삭제할 데이터의 갯수 세보기 
bicycle_data[(bicycle_data['대여 대여소번호']==bicycle_data['반납대여소번호']) & ( (bicycle_data['이용시간']==0) | (bicycle_data['이용거리']==0.0)) ].count()

# 갯수를 세보니 15003개 데이터가 필요없는 데이터로 확인되었음.
```

```python
# 좌표->주소 변환
from geopy.geocoders import Nominatim

def geocoding_reverse(lat_lng_str): 
    geolocoder = Nominatim(user_agent = 'South Korea', timeout=None)
    address = geolocoder.reverse(lat_lng_str)

    return address

address = geocoding_reverse('37.55495071,126.9108353')
address2 = geocoding_reverse('37.52464294,126.9234238')

print(address)
```

```python
rentalshop_data['address'] = None
```

```python
rentalshop_data['dong'] = None
```

```python
# 위도 경도로 주소(제대로 된)와 동 들고오기
for i, j in rentalshop_data.iterrows():

    print(i)
    g = geocoding_reverse((j['Unnamed: 4'], j['Unnamed: 5']))
    

    rentalshop_data['address'][i] = g[0]
    print(g[0])
    
    
    a_list = g[0].split(',')
    for c in range(1,len(a_list)):
        if a_list[c][-1] == '구' :
            if a_list[c-1][-1] == '동':
                dong = a_list[c-1].strip()
            else:
                dong ='없음'

    print('동:',dong)
    rentalshop_data['dong'][i] = dong

```

```python
# 불일치 대여소 번호 확인하기
li = []
for i in range(len(bicycle_data)):
    li.append(bicycle_data['대여 대여소번호'][i])
    li.append(bicycle_data['반납대여소번호'][i])

li1= list(set(li))
rentalshop_data.reset_index(inplace=True)
li2 = rentalshop_data['대여소번호'].to_list()
li2 = list(set(li2))
li3 = list(set(li1) -set(li2))
li4 = li3 + list(set(li2)-set(li1))
df = pd.DataFrame({'대여소번호':li4})

#df.to_csv('불일치 대여소 번호.csv', encoding='CP949')
```

```python
컬럼이름변경
return_stop = rentalshop_data[['대여소번호','소재지(위치)','Unnamed: 4','Unnamed: 5', 'Unnamed: 8', 'address', 'dong']]
return_stop.columns=['반납소번호', '반납 소재지','반납소 위도','반납소 경도', '반납소 보관수', '반납소 주소', '반납소 동']
```

```python
# 데이터 병합

df3 = pd.merge(bicycle_data, rental_stop, left_on = '대여 대여소번호', right_on='대여소번호', how='left')
df4 = pd.merge(df3, return_stop, left_on = '반납대여소번호', right_on='반납소번호', how='left')
```

```python
df4.drop(['대여소번호', '반납소번호'], axis=1, inplace=True)
```

```python
# 수리 지점 제거
idx = []
for i in range(len(df)):
    idx_num = df4.index[(df4['대여 대여소번호'] == df['대여소번호'][i])]
    if len(idx_num)> 0:
        for j in range(len(idx_num)):
            idx.append(idx_num[j])
for i in range(len(df)):
    idx_num = df4.index[(df4['반납대여소번호'] == df['대여소번호'][i])]
    if len(idx_num)> 0:
        for j in range(len(idx_num)):
            idx.append(idx_num[j])
df4 = df4.drop(idx)
```
