```python
# 대여이력 정보
import pandas as pd

bicycle_data = pd.read_csv("bicycle_data.csv", encoding='CP949', index_col=0)
bicycle_data.head(5)
```

```python
li = []
for i in range(len(bicycle_data)):
    li.append(bicycle_data['대여 대여소번호'][i])
    li.append(bicycle_data['반납대여소번호'][i])

li1= list(set(li))
rentalshop_data.reset_index(inplace=True)
li2 = rentalshop_data['대여소번호'].to_list()
li2 = list(set(li2))
li3 = list(set(li1) -set(li2))
li4 = li3 + list(set(li2)-set(li1))
df = pd.DataFrame({'대여소번호':li4})
#df.to_csv('불일치 대여소 번호.csv', encoding='CP949')
```

```python
return_stop = rentalshop_data[['대여소번호','소재지(위치)','Unnamed: 4','Unnamed: 5', 'Unnamed: 8', 'address', 'dong']]
return_stop.columns=['반납소번호', '반납 소재지','반납소 위도','반납소 경도', '반납소 보관수', '반납소 주소', '반납소 동']
return_stop
```

```python
rental_stop = rentalshop_data[['대여소번호','소재지(위치)','Unnamed: 4','Unnamed: 5','Unnamed: 8', 'address', 'dong']]
rental_stop.columns=['대여소번호', '대여 소재지','대여소 위도','대여소 경도','대여소 보관수', '대여소 주소', '대여소 동']
rental_stop
```

```python
df3 = pd.merge(bicycle_data, rental_stop, left_on = '대여 대여소번호', right_on='대여소번호', how='left')

df4 = pd.merge(df3, return_stop, left_on = '반납대여소번호', right_on='반납소번호', how='left')
```

```python
df4.drop(['대여소번호', '반납소번호'], axis=1, inplace=True)
```

```python
df4
```

```python
# 수리 지점 제거
idx = []
for i in range(len(df)):
    idx_num = df4.index[(df4['대여 대여소번호'] == df['대여소번호'][i])]
    if len(idx_num)> 0:
        for j in range(len(idx_num)):
            idx.append(idx_num[j])
for i in range(len(df)):
    idx_num = df4.index[(df4['반납대여소번호'] == df['대여소번호'][i])]
    if len(idx_num)> 0:
        for j in range(len(idx_num)):
            idx.append(idx_num[j])
df4 = df4.drop(idx)
```
