```python
##사업체수
import pandas as pd

f = pd.read_table("./_census_reqdoc_1659063002585//original//11_2019년_산업분류별(10차_대분류)_사업체수.txt",sep='^',header=None)
f = f.fillna(0)
f = f.iloc[:,1:]
f.iloc[:,0] = f.iloc[:,0].map(str)
f['읍면동코드'],f['일련번호'] = map(f.iloc[:,0].str.slice,[0,7],[7,12])
#print(f)

d = pd.read_excel("./ref_code//original//adm_code.xls",sheet_name='2019년6월')
d = d[d.iloc[:,0]=='11'] #서울특별시 정보만 추출
d = d.iloc[:,2:] #시도코드, 시도명칭 삭제
#print(d)

g = pd.read_excel("./ref_code//original//statistics_code.xls",sheet_name='집계구')
g = g.iloc[:,2:]
g = g[g.iloc[:,1].str.contains("cp_bnu_")] #원하는 사업체수 코드를 가진 row만 추출
g = g.tail(20)
#print(g)


#데이터 병합
merge1 = pd.merge(f,d,left_on='읍면동코드',right_on=d.iloc[:,2],how='left')
merge1 = merge1.iloc[:,[0,2,4,6,8]]
merge2 = pd.merge(f,g,left_on=f.iloc[:,1],right_on=g.iloc[:,1],how='left')
merge2 = merge2.iloc[:,[1,6]]
sa_up_chae_soo = pd.concat([merge1,merge2],axis=1,join='inner')
sa_up_chae_soo = sa_up_chae_soo.iloc[:,[1,2,3,4,6]]
names = ['사업체수','일련번호','구이름','동이름','산업종류']
sa_up_chae_soo.columns = names
names = names[1:]+names[:1]
sa_up_chae_soo = sa_up_chae_soo[names]
#print(sa_up_chae_soo)


sucs = sa_up_chae_soo.groupby(['동이름'])['사업체수'].transform(max) == sa_up_chae_soo['사업체수']
sucs2 =  sa_up_chae_soo[sucs]
sucs2

#sucs2.to_csv("C://Datacampus_project//사업체수max.csv", index = False)
```

```python
##종사자수
import pandas as pd

f = pd.read_table("./_census_reqdoc_1659063002585//original//11_2019년_산업분류별(10차_대분류)_종사자수.txt",sep='^',header=None)
f = f.fillna(0)
f = f.iloc[:,1:]
f.iloc[:,0] = f.iloc[:,0].map(str)
f['읍면동코드'],f['일련번호'] = map(f.iloc[:,0].str.slice,[0,7],[7,12])
#print(f)

d = pd.read_excel("./ref_code//original//adm_code.xls",sheet_name='2019년6월')
d = d[d.iloc[:,0]=='11'] #서울특별시 정보만 추출
d = d.iloc[:,2:] #시도코드, 시도명칭 삭제
#print(d)

g = pd.read_excel("./ref_code//original//statistics_code.xls",sheet_name='집계구')
g = g.iloc[:,2:]
g = g[g.iloc[:,1].str.contains("cp_bem_")] #원하는 종사자수 코드를 가진 row만 추출
g = g.tail(20)
#print(g)


#데이터 병합
merge1 = pd.merge(f,d,left_on='읍면동코드',right_on=d.iloc[:,2],how='left')
merge1 = merge1.iloc[:,[0,2,4,6,8]]
merge2 = pd.merge(f,g,left_on=f.iloc[:,1],right_on=g.iloc[:,1],how='left')
merge2 = merge2.iloc[:,[1,6]]
jong_sa_ja_soo = pd.concat([merge1,merge2],axis=1,join='inner')
jong_sa_ja_soo = jong_sa_ja_soo.iloc[:,[1,2,3,4,6]]
names = ['종사자수','일련번호','구이름','동이름','산업종류']
jong_sa_ja_soo.columns = names
names = names[1:]+names[:1]
jong_sa_ja_soo = jong_sa_ja_soo[names]
#print(jong_sa_ja_soo)



##grouping by Dong name
jsjs = jong_sa_ja_soo.groupby(['동이름'])['종사자수'].transform(max) == jong_sa_ja_soo['종사자수']
jsjs2 =  jong_sa_ja_soo[jsjs]
jsjs2['산업종류'].value_counts()
jsjs2.to_csv("C://Datacampus_project//종사자수max.csv", index = False)
```
