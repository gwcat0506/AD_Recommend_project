```python
df4 = pd.read_csv('Bike2.csv', encoding='utf8', index_col=0)
df4.head(3)
```

```python
list_dong = df4['동'].unique().tolist()
df4['동'] = df4['동'].map(lambda x: list_dong.index(x))
list_st = df4['대여소근처역'].unique().tolist()
df4['대여소근처역'] = df4['대여소근처역'].map(lambda x: list_st.index(x))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20,20))
heat1 = sns.heatmap(data=df4.corr(), annot=True, fmt='.2f', linewidths=.5,
           cmap='RdYlBu_r', vmin=-1, vmax = 1, cbar_kws={"shrink":.5})
heat1_fig = heat1.get_figure()

heat1_fig.savefig('heatmap1.jpg')
```

```python
plt.savefig('heatmap1.png')
```

```python
import numpy as np

fig, ax = plt.subplots(figsize=(20, 20))

mask = np.zeros_like(df4.corr(), dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

heat2 = sns.heatmap(data=df4.corr(), annot=True, mask=mask, fmt='.2f', linewidths=.5,
           cmap='RdYlBu_r', vmin=-1, vmax = 1, cbar_kws={"shrink": .5})

heat2_fig = heat2.get_figure()

heat2_fig.savefig('heatmap2.jpg')
```

```python
X = df4.drop(['광고'], axis=1)
y = df4['광고']
```

```python
import statsmodels.api as sm

df4['대여일자'] = df4['대여일자'].astype(str)
df4['대여일자'] = df4['대여일자'].str.replace(pat=r'[^\w]', repl=r'', regex=True)
df4['대여일자'] = df4['대여일자'].astype(int)

model = sm.OLS(y, X).fit()
model.summary()
```
