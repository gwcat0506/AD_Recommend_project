```python
# Module
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
from sklearn import tree
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
from category_encoders import OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline
from xgboost import XGBClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import RidgeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.dummy import DummyClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
import lightgbm as lgb
```

```python
# Data Processing
df = pd.read_csv('Bike2.csv', encoding='utf8', index_col=0)
list_dong = df['동'].unique().tolist()
df['동'] = df['동'].map(lambda x: list_dong.index(x))
list_st = df['대여소근처역'].unique().tolist()
df['대여소근처역'] = df['대여소근처역'].map(lambda x: list_st.index(x))
df['대여일자'] = df['대여일자'].astype(str)
df['대여일자'] = df['대여일자'].str.replace(pat=r'[^\w]', repl=r'', regex=True)
df['대여일자'] = df['대여일자'].astype(int)
```

```python
X = df.drop(['광고'], axis=1)
Y = df['광고']

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=0)

X_features =['대여일자', '대여시간', '대여소번호', '대여구분코드', '성별', '연령대코드', '이용건수', '운동량',
       '이동거리(M)', '이용시간(분)', '주말', '요일', '대여소근처역', '소재지', '동', '산업종류']
Y_target = [29, 12, 28,  8, 27, 22,  7,  1,  2, 25, 26, 13,  0, 23, 14, 20, 18,
       19,  3, 24,  6, 15, 16, 17, 21, 11,  4,  5, 10,  9]
```

```python
# Decision Tree Classifier
dt_clf = DecisionTreeClassifier(random_state=0)
dt_clf.fit(X_train, y_train)
```

```python
pred = dt_clf.predict(X_test)
accuracy = accuracy_score(y_test, pred)
print('Decision Tree 예측 정확도 : {0:.4f}'.format(accuracy))
print('\nDecisionTreeClassifier 기본 하이퍼파라미터:\n', dt_clf.get_params())
```

```python
y_pred = dt_clf.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
cross_val_score(dt_clf, X, Y, cv=5)
```

```python
# ExtraTreesClassifier
xtree = ExtraTreesClassifier(n_estimators=5, random_state=0)
xtree.fit(X_train, y_train)
```

```python
print("훈련 세트 정확도: {:.3f}".format(xtree.score(X_train, y_train)))
print("테스트 세트 정확도: {:.3f}".format(xtree.score(X_test, y_test)))
```

```python
y_pred = xtree.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# Naive Bayes
scaler = StandardScaler()
scaler.fit(X_train, y_train)
X_train_transformed = scaler.transform(X_train)
X_test_transformed = scaler.transform(X_test) 
gnb = GaussianNB()
gnb.fit(X_train_transformed, y_train) 
y_pred = gnb.predict(X_test_transformed)
accuracy = accuracy_score(y_test, y_pred)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print('Naive Bayes 예측 정확도 : {0:.4f}'.format(accuracy))
print('\nNaive Bayes 기본 하이퍼파라미터:\n', gnb.get_params())
```

```python
# LinearDisciminantAnalysis
lda = LinearDiscriminantAnalysis()
lda.fit(X_train,y_train)
```

```python
lda.fit(X_train, y_train)
score_train_pre = lda.score(X_train, y_train)
score_test_pre = lda.score(X_test, y_test)
print('{:.3f}'.format(score_train_pre))
print('{:.3f}'.format(score_test_pre))
```

```python
y_pred = lda.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print('LinearDisciminantAnalysis 예측 정확도 : {0:.4f}'.format(accuracy))
print('\nLinearDisciminantAnalysis 기본 하이퍼파라미터:\n', lda.get_params())
```

```python
# Ridge Classifier
rcf = RidgeClassifier().fit(X_train, y_train)
rcf.score(X_train, y_train)
```

```python
rcf.fit(X_train, y_train)
score_train_pre = rcf.score(X_train, y_train)
score_test_pre = rcf.score(X_test, y_test)
y_pred = rcf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print('{:.3f}'.format(score_train_pre))
print('{:.3f}'.format(score_test_pre))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print('Ridge Classifier 예측 정확도 : {0:.4f}'.format(accuracy))
print('\nRidge Classifier 기본 하이퍼파라미터:\n', rcf.get_params())
```

```python
# Logistic Regression
lgr = LogisticRegression()
lgr.fit(X_train, y_train)
score_train_pre = lgr.score(X_train, y_train)
score_test_pre = lgr.score(X_test, y_test)
pred = lgr.predict(X_test)
print('{:.3f}'.format(score_train_pre))
print('{:.3f}'.format(score_test_pre))
print(classification_report(y_test, pred))
```
