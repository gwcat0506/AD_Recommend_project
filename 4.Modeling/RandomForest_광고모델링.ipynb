```python
# 학습시킬 데이터 로드

import pandas as pd

df = pd.read_csv('C://Users//gwcat//Downloads//Bike2.csv')
```

```python
df
```

```python
del df['Unnamed: 0']
del df['대여일자']
```

```python
# one-hot 인코딩 하니 컬럼이 너무 많아 메모리 에러..
# 동과 대여소근처역을 직접 숫자로 mappling 해서 돌리기
```

```python
# 동 mappling
list_dong = df['동'].unique().tolist()
df['동'] = df['동'].map(lambda x: list_dong.index(x))
```

```python
# 대여소근처역 mappling
list_station = df['대여소근처역'].unique().tolist()
df['대여소근처역'] = df['대여소근처역'].map(lambda x: list_station.index(x))
```

```python
X = df.drop(['광고'], axis=1)
Y = df['광고']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)
```

```python
#기본적인 randomforest모형

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score # 정확도 함수

clf = RandomForestClassifier(n_estimators=20, max_depth=5,random_state=0)
clf.fit(X_train,Y_train)

predict1 = clf.predict(X_test)
print("accuracy_score =",accuracy_score(Y_test,predict1))
```
