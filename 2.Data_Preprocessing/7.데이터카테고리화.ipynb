### 1. 데이터 로드

```python
import pandas as pd

data = pd.read_csv('./final_final.csv')
data
```

```python
data.info()
```

### 2. 데이터 전처리

```python
# 필요없는 컬럼 삭제 
del data['번호']
del data['대여소명']
del data['위도']
del data['경도']
del data['산업종류']
```

```python
data.rename(columns={'대여시간':'대여시각'})
```

```python
# 운동량 전처리
data['운동량'].replace({'\\N':0.00},inplace=True)
data.astype({'운동량':'float'})
```

```python
# 탄소량 전처리 
data['탄소량'].replace({'\\N':0.00},inplace=True)
data.astype({'탄소량':'float'})
```

```python
# 최종 16개 데이터 
data.info()
```

### 3. 데이터 카테고리화 ###

```python
# 1. 대여구분코드
data['대여구분코드'].replace({'정기권':0, '일일권':1, '일일권(비회원)':2, '단체권':3}, inplace=True)
data['대여구분코드'].value_counts()
```

```python
# 2. 성별
data['성별'].replace({'M':0,'N':1,'F':2}, inplace=True)
data['성별'].value_counts()
```

```python
# 3. 연령대코드
data['연령대코드'].replace({'N':0, '10대':1, '20대':2, '30대':3,
'40대':3, '50대':4, '60대':4, '70대':5},
inplace=True)
data['연령대코드'].value_counts()
# 구분 기준 https://www.gosims.go.kr/hh/hh001/retrieveLfcySearch.do
```

```python
# 4. 주말
data['주말'].replace({'평일':0,'주말':1},inplace=True)
data['주말'].value_counts()
```

```python
# 5. 요일
data['요일'].replace({'월':0,'화':1,'수':2,'목':3,'금':4,'토':5,'일':6},inplace=True)
data['요일'].value_counts()
```

```python
# 6. 소재지

data['소재지'].replace({'강서구':0,'송파구':1,'영등포구':2,'노원구':3,'양천구':4,'마포구':5,'광진구':6,
'강동구':7,'구로구':8,'성동구':9,'동대문구':10,'종로구':11,'서초구':12,'강남구':13,'중랑구':14,
'관악구':15,'은평구':16,'성북구':17,'중구':18,'용산구':19,'서대문구':20,'동작구':21,'도봉구':22,
'금천구':23,'강북구':24},inplace=True)
data['소재지'].value_counts()
```
